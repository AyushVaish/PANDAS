Grouping and aggregating

```python
!gdown 1E3bwvYGf1ig32RmcYiWc0IXPN-mD_bI_
```

```python
mck = pd.read_csv('mckinsey.csv')
```

```python
mck
```

```python
x = mck.groupby('continent')[['population']].max()
x
```

```python
df2 = x.reset_index()
df2
```

```python
mck
```

```python
pd.merge(df2,mck,on=["continent",'population'])
```

```python
mck.groupby('continent')['population'].aggregate(['max','min','count','mean'])
```

```python
mck.groupby('continent')['population'].aggregate(['max','min','count','mean']).reset_index()
```

```python
df3 = mck.groupby('continent')[['population','gdp_cap']].aggregate(['max','min'])
df3
```

```python
result = mck.groupby('continent')[['population','gdp_cap']].aggregate({'population':'max','gdp_cap':'mean'})
```

```python
result.rename(columns={'population':'pop_max','gdp_cap':'avg_gcap'})
```

```python
temp = mck.query('year>2000')
df4 = temp.groupby(['continent','year'])[['population','gdp_cap']].aggregate(['max','min'])
df4
```

```python
df4.loc[("Europe",2002),("population","max")]
```

```python
df4.reset_index()
```
