Concat and Merge

```python
df1 = pd.DataFrame({"name": ['tobey', 'peter', 'chris', 'pratt'],'chem_score':[10, 9, 8, 7], 'phy_score':[7, 8, 9, 10], 'Hindi_score':[9, 9, 9, 9]})
df2 = pd.DataFrame({"name": ['chris', 'pratt', 'andrew', 'tom'],'chem_score':[10, 10, 10, 9], 'maths_score':[6, 6, 7, 9], 'eng_score':[9, 10, 10, 9]})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2]).iloc[0].count()
```

```python
pd.concat([df1,df2]).count(axis =0)
```

```python
pd.concat([df1,df2]).info()
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,on='name',how='inner')
```

```python
pd.merge(df1,df2,on='name',how='inner').mean(axis=1).iloc[0]
```
